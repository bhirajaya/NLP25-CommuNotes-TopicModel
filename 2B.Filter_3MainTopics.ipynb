```python
import pandas as pd
import numpy as np
import os
import re
```

```python
path = '2.politics_notes/'
df_nm = pd.read_excel(path+'nonmis_politics (1).xlsx')
df_mis = pd.read_excel(path+'mislead_politics (1).xlsx')
df_nm['Mislead_NonMislead'] = ['NonMislead']*df_nm.shape[0]
df_mis['Mislead_NonMislead'] = ['Mislead']*df_mis.shape[0]
df = pd.concat([df_nm,df_mis])
```

```python
def filter_project_25(df):
  # Regular expression to match variations of "project 25"
  pattern = r"project\s*2025|proj\s*2025|p2025"  # Example: project 25, proj 25, p25
  # Use the str.contains method with the regex pattern and case-insensitive flag
  filtered_df = df[df['Document'].str.contains(pattern, case=False, regex=True)]
  return filtered_df

def filter_kamala_harris(df):
  # Regular expression to match variations of "Kamala" or "Harris"
  pattern = r"kamala|harris|Kamala|Harris"  # Example: Kamala, Harris, kamala, harris
  # Use the str.contains method with the regex pattern and case-insensitive flag
  filtered_df = df[df['Document'].str.contains(pattern, case=False, regex=True)]
  return filtered_df


def filter_immigration(df):
  # Regular expression to match variations of "immigration"
  pattern = r"immigration|immigrant|immigrants|immigrate|immigrated|immigrating"  # Example: immigration, immigrant, immigrants, immigrate etc.
  # Use the str.contains method with the regex pattern and case-insensitive flag
  filtered_df = df[df['Document'].str.contains(pattern, case=False, regex=True)]
  return filtered_df
```

```python
p25_df = filter_project_25(df)
p25_df
```

```python
kamala_df = filter_kamala_harris(df)
kamala_df
```

```python
immi_df = filter_immigration(df)
immi_df
```
