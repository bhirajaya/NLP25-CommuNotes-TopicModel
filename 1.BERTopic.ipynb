```python
import pandas as pd
# https://maartengr.github.io/BERTopic/getting_started/quickstart/quickstart.html#quick-start
from bertopic import BERTopic
# from sklearn.datasets import fetch_20newsgroups

from sentence_transformers import SentenceTransformer
from umap import UMAP
from bertopic.representation import KeyBERTInspired

# To display full column content, set it to None
pd.set_option('display.max_colwidth', None)
```

```python
path = 'topic_model_results/'
```

### Read Data

```python
df = pd.read_csv('raw_data_samples/sample_english_notes-00000.tsv', sep='\t')
df
```

```python
mis_doc = df[df['classification']=='MISINFORMED_OR_POTENTIALLY_MISLEADING']['summary']
nm_doc = df[df['classification']=='NOT_MISLEADING']['summary']
```

### Not Misleading

```python
# from bertopic.representation import KeyBERTInspired

# Fine-tune your topic representations
representation_model = KeyBERTInspired()
topic_model = BERTopic(representation_model=representation_model)
topics, probs = topic_model.fit_transform(nm_doc)
```

```python
topic_model.save(path+"0.TopicModel_NonMisLead", serialization="pickle")
```

```python
topic_model.visualize_topics()
```

```python
df_doc_nm = topic_model.get_document_info(nm_doc)
df_doc_nm.to_csv(path+'1A.Results_NonMislead.csv')
```

```python
df_doc_nm.head()
```

```python
df_doc_nm_gb = df_model.groupby(['Topic','Name']).count()
df_doc_nm_gb.to_csv(path+'1B.GroupedResults_NonMisLead.csv')
df_doc_nm_gb
```

```python
# Hierarchical Grouping
hier_nm = topic_model.hierarchical_topics(nm_doc)
topic_model.visualize_hierarchy(hierarchical_topics=hier_nm)
```

```python
hier_nm.to_csv(path+'2.Hierarchical_NonMislead.csv')
```

```python
# df_doc_nm_gb = df_doc_nm.groupby(['Topic','Name']).count().reset_index()

# hier_nm = model_nm.hierarchical_topics(nm_doc)
# group topic that distance less than 1
# more than 1 topic is kinda too broad
hier_nm = hier_nm.sort_values('Distance', ascending=False)
hier_nm_filter = hier_nm[hier_nm['Distance']<1]
hier_nm_filter = hier_nm_filter.reset_index()

ll = []

# loop each topic
for i in range(len(df_doc_nm_gb)):
  # print(i)
  topic = df_doc_nm_gb['Topic'][i]
  is_found = False
  # loop each hier that less than 1
  for j in range(len(hier_nm_filter)):
    # for topics list
    topic_l = hier_nm_filter['Topics'][j]
    if topic in topic_l:
      is_found = True
      ll.append([topic,hier_nm_filter['Parent_ID'][j],hier_nm_filter['Parent_Name'][j],hier_nm_filter['Topics'][j]])
      break
  if not is_found:
    ll.append([topic,'','',''])

hier_merge = pd.DataFrame(ll, columns=['Topic','Parent_ID','Parent_Name','Topics'])
hier_merge
# df_doc_nm_gb_hier = df_doc_nm[['Topic','Name','Document']].merge(hier_merge, on='Topic')
# df_doc_nm_gb_hier.groupby('Parent_Name').agg({'Document':'sum','Name':'count'}).reset_index()
```

```python
df_doc_nm_gb_hier_x = df_doc_nm.merge(hier_merge, on='Topic')
df_doc_nm_gb_hier_x.to_csv(path+'3.Results_withHierParents_NonMislead.csv')
```
